According to a study by Bark, the company who sells BarkBox subscription boxes for your dogs, "one in ten (11%)" dog owners in America manage a social media account for their canine companion(s) ("New BarkBox Study", 2017). If you are one of these dog owners, then you might be interested in what can maximize exposure and interest. 

Look no further! As I have analyzed data from the WeRateDogs Twitter archive and found what can influence retweets and favorites.

> WeRateDogs is a humorous Twitter account that rates images of dogs submitted by their followers. It was created in 2015 and has amassed a following of 9.3 million.    


### Analysis Questions
1. Which stage of the dogs in this dataset gets the most favorites?
2. Which stage of the dogs in this dataset gets the most retweets?
3. Is there a correlation between favorites and retweets?

Before I present the analysis and visuals, I want to clarify some terms:

> **Stage:** A dog classification of `doggo`, `pupper`, `floofer`, or `puppo`. These are terms whose definitions were created by WeRateDogs. Their definitions can be found in The Dogtionary (see next term).

> **The Dogtionary:** A dictionary of dog-related terms whose definitions were created by WeRateDogs. It can be found in their book #WeRateDogs.

> **Favorite:** A way that a Twitter user can engage with a Tweet that indicates they show interest in it.

> **Retweet:** When a Twitter user takes another users original Tweet and replies to it on their own feed. 

#### Question 1
Which stage of the dogs in this dataset gets the most favorites?

For this question I compared the total number of favorites that each stage received. From the visualization that was produced, we can see that dogs classified as a "puppo" receive the most interest from other users. Since a puppo is essentially the dog version of a human teenager, and people love pictures of dogs who are still learning how to behave, this insight makes sense. 

> The stage `puppo` gets the most favorites.

#### Question 2
Which stage of the dogs in this dataset gets the most retweets?

Now I am comparing the total number of retweets that each stage received. From the visualization that was produced, we can see that dogs classified as "doggo" are the most shared by other users. This insight was not exactly what I had expected, because if "puppos" get more favorites then one could think that they get more retweets as well. This prompted my next question.

> The stage `doggo` gets the most retweets.

#### Question 3
Is there a correlation between favorites and retweets?

This question required a simple scatter plot comparing the number of favorites a Tweet receives to the number of retweets is also received. This visualization shows a positive relationship between the two variables. As the number of favorites increase, the number of retweets also increase. However, we can also see that as these numbers grow, the number of Tweets decreases.

> There is a strong and positive correlation between `retweet_count` and `favorite_count`.

### Conclusion
If you want to maximze exposure and interest for your dog's social media account, keep these insights in mind:

- Pictures of your dog(s) going through their "teenager-like" phase (a puppo) could generate a lot of interest in the form of favorites or likes.



- Pictures of your dog(s) doing serious adult-like things (a doggo) could generate a lot of exposure in the form of retweets or shares. 



- There is a strong and positive relationship between retweets and favorites. 

Therefore, either of the two above types of pictures have the potential to generate both exposure and interest.
